#2) Transformation logarithmique

La transformation logarithmique est largement utilisée dans la recherche biomédicale et psychosociale pour traiter les données asymétriques. Si les données originales suivent une distribution log-normale ou approximativement normale, alors les données log-transformées suivent une distribution normale ou presque normale. Dans ce cas, la transformation logarithmique supprime ou réduit l'asymétrie. Malheureusement, les données issues de nombreuses études ne se rapprochent pas de la distribution log-normale et l'application de cette transformation ne réduit pas l'asymétrie de la distribution. En fait, dans certains cas, l'application de la transformation peut rendre la distribution plus asymétrique que les données d'origine.



```python
df["Log_Potatoes"] = df["Potatoes"].map(lambda i: np.log(i) if i > 0 else 0) 
```

```python
df["Log_Potatoes"]
```

```python

print(df['Potatoes'].skew()) # skew avant reglage
print(df['Log_Potatoes'].skew()) #skew apres relage
```
